```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn import datasets
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.metrics import confusion_matrix
```

```python
iris = datasets.load_iris()
```

```python
#7. Utiliza train/split y KNN para entrenar un modelo

#Hacemos split para sacar los 4 arrays que nos haran falta
X_train, X_test, Y_train, Y_test = train_test_split(iris.data ,iris.target ,  test_size = 0.5)
list = []
#Utilizamos el metodo KNN para entrenar el modelo
for i in range(1,50):
    knn = KNeighborsClassifier(n_neighbors= i)
    knn.fit(X_train, Y_train)
    pred = knn.predict(X_test)
    list.append(accuracy_score(Y_test, pred))
    
```

```python
#Mejor modelo KNN
print("El mejor modelo para KNN conseguido es con K = ", np.argmax(list)+1, " y nos da un % de ", 
      np.max(list)*100)
```

```python
#Peor modelo KNN
print("El peor modelo para KNN conseguido es con K = ", np.argmin(list)+1, " y nos da un % de ", 
      np.min(list)*100)
```

```python
#Matriz de confusion para mejor modelo de KNN
knn = KNeighborsClassifier(n_neighbors= np.argmax(list) + 1)
knn.fit(X_train, Y_train)
pred = knn.predict(X_test)
confusion_matrix(Y_test, pred)
```

```python
#Matriz de confusion para peor modelo de KNN
knn = KNeighborsClassifier(n_neighbors= np.argmin(list) + 1)
knn.fit(X_train, Y_train)
pred = knn.predict(X_test)
confusion_matrix(Y_test, pred)
```

```python
from sklearn import svm
```

```python
from sklearn.model_selection import cross_val_predict
```

```python
from sklearn import metrics
```

```python
#8. Cross validation

list_CV = [] #Cambiar iris.data y iris.target por X_test e Y_test ?
for i in range(1,50):
    clf = svm.SVC(kernel='linear', C=i)
    score = cross_val_predict(clf, iris.data, iris.target, cv=10)
    list_CV.append(score)
```

```python
#Mejor modelo CV
print("El mejor modelo para Cross validation conseguido es con K = ", np.argmax(list_CV)+1, " y nos da un % de ", 
      np.max(list_CV))
```

```python
#Peor modelo CV
print("El peor modelo para Croos validation conseguido es con K = ", np.argmin(list_CV)+1, " y nos da un % de ", 
      np.min(list_CV))
```

```python
#Matriz de confusion para mejor modelo CV
clf = svm.SVC(kernel='linear', C=np.argmax(list) + 1).fit(X_train, Y_train)
predicted = cross_val_predict(clf, iris.data, iris.target, cv=10)
confusion_matrix(iris.target, predicted)
```

```python
#Matriz de confusion para peor modelo CV
clf = svm.SVC(kernel='linear', C=np.argmin(list) + 1).fit(X_train, Y_train)
predicted = cross_val_predict(clf, iris.data, iris.target, cv=10)
confusion_matrix(iris.target, predicted)
```
